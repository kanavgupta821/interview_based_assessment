```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_nasa.csv')
```

```python
df
```

## Get all the Earth meteorites that fell before the year 2000

```python
df['year'] = pd.to_datetime(df['year'])
```

```python
meteorites = df[df['year'].dt.year < 2000]
meteorites
```

## Get all the earth meteorites co-ordinates who fell before the year 1970

```python
coordinates = df[df['year'].dt.year < 1970][['name','reclong','reclat', 'coordinates']]
```

```python
coordinates
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(coordinates['reclong'], coordinates['reclat'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorites Coordinates Before 1970')
plt.grid(True)
plt.show()
```

## Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
mass = df[df['mass'] > 10000]
```

```python
mass
```

```python
plt.figure(figsize=(8, 6))
plt.hist(mass['mass'], bins=10)
plt.xlabel('Mass (kg)*10^7')
plt.ylabel('Count of Meteorites')
plt.title('Earth Meteorites Mass > 10000kg')
plt.grid(True)
plt.show()
```
