```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Month'] = label_encoder.fit_transform(df['Month'])
df['VisitorType'] = label_encoder.fit_transform(df['VisitorType'])

# Separate features and target variables
X = df.drop(['Revenue', 'Weekend', 'Informational_Duration'], axis=1)
y_revenue = df['Revenue']
y_weekend = df['Weekend']
y_duration = df['Informational_Duration']

# Split the data into training and testing sets
X_train, X_test, y_revenue_train, y_revenue_test, y_weekend_train, y_weekend_test, y_duration_train, y_duration_test = train_test_split(X, y_revenue, y_weekend, y_duration, test_size=0.2, random_state=42)

# Train the Random Forest classifier/regressor
revenue_classifier = RandomForestClassifier()
revenue_classifier.fit(X_train, y_revenue_train)

weekend_classifier = RandomForestClassifier()
weekend_classifier.fit(X_train, y_weekend_train)

duration_regressor = RandomForestRegressor()
duration_regressor.fit(X_train, y_duration_train)

# Make predictions
revenue_predictions = revenue_classifier.predict(X_test)
weekend_predictions = weekend_classifier.predict(X_test)
duration_predictions = duration_regressor.predict(X_test)

# Evaluate the model
revenue_accuracy = accuracy_score(y_revenue_test, revenue_predictions)
weekend_accuracy = accuracy_score(y_weekend_test, weekend_predictions)
duration_mse = mean_squared_error(y_duration_test, duration_predictions)

# Print the evaluation results
print("Revenue Accuracy:", revenue_accuracy)
print("Weekend Accuracy:", weekend_accuracy)
print("Duration Mean Squared Error:", duration_mse)
```
