```python
# Top 10 Companies in terms of no. of vehicles present on road
top_ten = df['Make'].value_counts().sort_values(ascending=False).head(10).reset_index()
top_ten
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='index',y='Make',data=top_ten)
plt.xticks(rotation=90)
plt.xlabel('CAR BRANDS')
plt.ylabel('COUNT')
plt.show()
```

```python
# Model wise classification of Tesla cars
temp = df[df['Make'] == 'TESLA']
tesla = temp.groupby('Model').count()['DOL Vehicle ID'].reset_index()
tesla['count'] = tesla['DOL Vehicle ID']
tesla.drop(labels=['DOL Vehicle ID'],axis=1,inplace=True)
```

```python
tesla
```

```python
plt.figure(figsize=(10,10))
plt.pie(tesla['count'],labels=tesla['Model'],autopct='%0.1f%%')
plt.title('Tesla Cars Modelwise Breakage')
plt.show()
```

## Draw plots to show the distribution between city and electric vehicle type

```python
import matplotlib.pyplot as plt

# Group the data by city and electric vehicle type and calculate the count
group = df.groupby(['City', 'Electric Vehicle Type']).count()['VIN (1-10)'].unstack()
group.replace(np.nan,0,inplace=True)
group['Total_Cars'] = group['Battery Electric Vehicle (BEV)'] + group['Plug-in Hybrid Electric Vehicle (PHEV)']
top_five = group.sort_values('Total_Cars',ascending=False).head(5)             # Top five cities in terms of total electric cars
top_five
```

```python
import matplotlib.pyplot as plt

# Create a figure object and specify the figure size
fig, ax = plt.subplots(figsize=(8, 6))

# Plot a bar plot
top_five[['Battery Electric Vehicle (BEV)', 'Plug-in Hybrid Electric Vehicle (PHEV)']].plot(kind='bar', ax=ax)

# Set the plot title and labels
ax.set_title('Distribution of Electric Vehicle Types in Top Five Cities')
ax.set_xlabel('City')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```

```python
# Create a figure object and specify the figure size
fig, ax = plt.subplots(figsize=(8, 6))

# Plot a stacked bar plot
top_five[['Battery Electric Vehicle (BEV)', 'Plug-in Hybrid Electric Vehicle (PHEV)']].plot(kind='bar', stacked=True,ax=ax)

# Set the plot title and labels
ax.set_title('Distribution of Electric Vehicle Types and Total Count in Top Five Cities')
ax.set_xlabel('City')
ax.set_ylabel('Count')

# Display the plot
plt.show()
```
