```python
import requests
import pandas as pd
import numpy as np
import datetime as datetime
```

```python
# Fetch the JSON data from the URL
url = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'
response = requests.get(url)
data = response.json()
```

```python
data
```

```python
data['id']
data['url']
data['name']
```

```python
main_data = data['_embedded']['episodes']
main_data
```

```python
# Convert the JSON data to DataFrame
df = pd.DataFrame(main_data)
df
```

```python
# Modifying the dataframe according to our requirement
avg_rating = []
for i in df['rating']:
    if type(i) == dict:
        avg_rating.append(i['average'])
    else:
        avg_rating.append(np.nan)
        
medium_image = []
original_image = []
for i in df['image']:
    if type(i) == dict:
        medium_image.append(i['medium'])
        original_image.append(i['original'])
        
    else:
        medium_image.append(np.nan)
        original_image.append(np.nan)
summary = []
for i in df['summary']:
    summary.append(i.replace('<p>','').replace('</p>',''))
```

```python
summary
```

```python
df['avg_rating'] = avg_rating
df['medium_image_link'] = medium_image
df['original_image_link'] = original_image
df['summary'] = summary
```

```python
df
```

```python
df.drop(labels=['rating','airstamp','image','_links'],axis=1,inplace=True)
```

```python
df.info()
```

```python
# converting the datatypes
df['runtime'] = df['runtime'].astype(float)
df['airdate'] = pd.to_datetime(df['airdate'])
for i in range(len(df['airtime'])):
    df['airtime'][i] = datetime.datetime.strptime(df['airtime'][i], '%I:%M %p').strftime('%I:%M %p')
```

```python
df
```

```python
# Export DataFrame to CSV file
output_file = 'westworld.csv'
df.to_csv(output_file, index=False)
```
